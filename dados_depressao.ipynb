## Sheet 3 - Proporção de diagnosticos de depressão por nivel de instrução

```python
def tratamento_dataset(dataset,colunas_remocao,nomes_colunas): 
    dataset.dropna(inplace=True)
    dataset.drop(columns=dataset.columns[colunas_remocao],inplace=True)

    dataset.columns = nomes_colunas

    dataset.drop(dataset.iloc[0:3].index,inplace=True)
    selecao = (dataset['Estado'] == 'Norte') | (dataset['Estado'] == 'Nordeste') | (dataset['Estado'] == 'Sudeste') | (dataset['Estado'] == 'Centro-Oeste') | (dataset['Estado'] == 'Sul')
    dataset.drop(dataset[selecao].index,inplace=True)
    dataset.reset_index(drop=True,inplace=True)

    dataset.loc[0:7, 'Região'] = 'Norte'
    dataset.loc[7:16, 'Região'] = 'Nordeste'
    dataset.loc[16:20, 'Região'] = 'Sudeste'
    dataset.loc[20:23, 'Região'] = 'Sul'
    dataset.loc[23:, 'Região'] = 'Centro-Oeste'

    return dataset
    

tab3_44_corRaca = pd.read_excel('dados/3.44.xls',sheet_name='cor_ou_raca')
colunas_remocao = [2,3,5,6,8,9,11,12]
nomes_colunas = ['Estado','Total','Branca','Preta','Parda']
tab3_44_corRaca = tratamento_dataset(tab3_44_corRaca,colunas_remocao,nomes_colunas)
tab3_44_corRaca.sample(5)
```

```python
tab3_44_posOcup = pd.read_excel('dados/3.44.xls',sheet_name='pos ocup')
colunas_remocao = [2,3,5,6,8,9,11,12]
nomes_colunas = ['Estado','Total','Trabalhando ocupado','Trabalhando desocupado','Desempregados']
tab3_44_posOcup = tratamento_dataset(tab3_44_posOcup,colunas_remocao,nomes_colunas)
tab3_44_posOcup.sample(5)
```

```python
tab3_44_rendimento = pd.read_excel('dados/3.44.xls',sheet_name='rendimento')
colunas_remocao = [2,3,5,6,8,9,11,12,13,15,16,18,19,21,22,24,25]
nomes_colunas = ['Estado','Total','Sem rendimento e ate 1/4 do salario minimo','Mais de 1/4 a 1/2 do salario minimo','Mais de 1/2 a 1 salario minimo','Mais de 1 a 2 salarios minimos', 'Mais de 2 a 3 salarios minimos', 'Mais de 3 a 5 salarios minimos', 'Mais de 5 salarios minimos']
tab3_44_rendimento = tratamento_dataset(tab3_44_rendimento,colunas_remocao,nomes_colunas)
tab3_44_rendimento.sample(5)
```

# Tratando dados capitais

```python
def tratamento_dataset_cidade(dataset,colunas_remocao,nomes_colunas): 
    dataset.dropna(inplace=True)
    dataset.drop(columns=dataset.columns[colunas_remocao],inplace=True)

    dataset.columns = nomes_colunas

    dataset.drop(dataset.iloc[0:1].index,inplace=True)
    selecao = (dataset['Estado/Cidade'] == 'Norte (1)') | (dataset['Estado/Cidade'] == 'Nordeste') | (dataset['Estado/Cidade'] == 'Sudeste (1)') | (dataset['Estado/Cidade'] == 'Centro-Oeste (1)') | (dataset['Estado/Cidade'] == 'Sul (1)')
    dataset.drop(dataset[selecao].index,inplace=True)
    dataset.reset_index(drop=True,inplace=True)

    dataset.loc[0:7, 'Região'] = 'Norte'
    dataset.loc[7:16, 'Região'] = 'Nordeste'
    dataset.loc[16:20, 'Região'] = 'Sudeste'
    dataset.loc[20:23, 'Região'] = 'Sul'
    dataset.loc[23:, 'Região'] = 'Centro-Oeste'

    return dataset
```

```python
tab3_44_sexCap = pd.read_excel('dados/3.44.xls',sheet_name='sexo (cap)')
colunas_remocao = [2,3,5,6,8,9]
nomes_colunas = ['Estado/Cidade','Diagnostico Depressão Total','Masculino','Feminino']
tab3_44_sexCap = tratamento_dataset_cidade(tab3_44_sexCap,colunas_remocao,nomes_colunas)
```

```python
tab3_44_corRaca = pd.read_excel('dados/3.44.xls',sheet_name='cor_ou_raca')
colunas_remocao = [2,3,5,6,8,9,11,12]
nomes_colunas = ['Estado','Total','Branca','Preta','Parda']
tab3_44_corRaca = tratamento_dataset(tab3_44_corRaca,colunas_remocao,nomes_colunas)
tab3_44_corRaca.sample(5)
```
