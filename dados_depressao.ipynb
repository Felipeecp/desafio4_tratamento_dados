# Tab 3.44
## Sheet 1 - Proporção de diagnosticos de depressão de acordo com o sexo

```python
tab3_44_sexo = pd.read_excel('dados/3.44.xls')
```

### Removendo colunas desnecessárias

```python
tab3_44_sexo.dropna(inplace=True)
tab3_44_sexo.drop(columns=['Unnamed: 2','Unnamed: 3','Unnamed: 5','Unnamed: 6','Unnamed: 8','Unnamed: 9'],inplace=True)
```

### Renomaendo colunas

```python
colunas = ['Estado','Diagnostico Depressão Total', 'Diagnostico Depressão Masculino', 'Diagnostico Depressão Feminino']
tab3_44_sexo.columns = colunas
```

```python
tab3_44_sexo.head(5)
```

### Removendo linhas com dados totais do brasil e de regiões, mantando apenas as colunas de cada estado

```python
tab3_44_sexo.drop(tab3_44_sexo.iloc[0:3].index,inplace=True)
selecao = (tab3_44_sexo['Estado'] == 'Norte') | (tab3_44_sexo['Estado'] == 'Nordeste') | (tab3_44_sexo['Estado'] == 'Sudeste') | (tab3_44_sexo['Estado'] == 'Centro-Oeste') | (tab3_44_sexo['Estado'] == 'Sul')
tab3_44_sexo.drop(tab3_44_sexo[selecao].index,inplace=True)
tab3_44_sexo.reset_index(drop=True,inplace=True)
```

```python
tab3_44_sexo.reset_index(drop=True,inplace=True)
```

```python
tab3_44_sexo.loc[0:7, 'Região'] = 'Norte'
tab3_44_sexo.loc[7:16, 'Região'] = 'Nordeste'
tab3_44_sexo.loc[16:20, 'Região'] = 'Sudeste'
tab3_44_sexo.loc[20:23, 'Região'] = 'Sul'
tab3_44_sexo.loc[23:, 'Região'] = 'Centro-Oeste'
```

### Corrigindo tipo de colunas numericas

```python
tab3_44_sexo.dtypes
```

```python
colunas_numericas = ['Diagnostico Depressão Total', 'Diagnostico Depressão Masculino', 'Diagnostico Depressão Feminino']
tab3_44_sexo[colunas_numericas] = tab3_44_sexo[colunas_numericas].apply(pd.to_numeric, errors='coerce')
tab3_44_sexo.head()
```
